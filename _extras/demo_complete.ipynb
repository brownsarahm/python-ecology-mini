# Grouped Data

Sometimes we want to split the data by the values in some category and look at statistics 

````python
# remember how the data is organized
surveys_df.head()
````

````python
# %load ../code/grouped.py
# Group data by sex
# complete the next row to group data by sex
grouped_data = surveys_df.groupby('sex')
# Compute the summary statistics for all numeric columns by sex
grouped_data.describe()
````

````python
# Provide the mean for each numeric column by sex
grouped_data.mean()
````

````python
# %load ../code/grouped_challenge
# How many recorded individuals are female `F` and how many male `M`
grouped_data['record_id'].count()
````

````python
# What happens when you group by two columns using the following syntax and
#    then grab mean values:
grouped_data2 = surveys_df.groupby(['plot_id','sex'])
grouped_data2.mean()
````

````python
# Summarize weight values for each plot in your data. HINT: you can use the
#  following syntax to only create summary statistics for one column in your data
by_plot = surveys_df.groupby('plot_id')
by_plot['weight'].describe()
````

We can now combine the things we've seen together

````python
# %load ../code/species_counts
# Count the number of samples by species_id
# HINT: use record_id as a column to count
species_counts = surveys_df.groupby('species_id')['record_id'].count()
print(species_counts)
````

````python
# 1. Create a plot of average weight across all species per plot.
by_plot_sex = surveys_df.groupby(['plot_id','sex'])
plot_sex_weight = by_plot_sex['weight'].mean()
plot_sex_species = surveys_df.groupby(['plot_id','species_id'])['record_id'].count()
spc = plot_sex_weight.unstack()
s_plot = spc.plot(kind='bar',stacked=True,title=" Mean weight by plot and sex")
s_plot.set_ylabel("Weight")
s_plot.set_xlabel("Plot");
````
